```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.append("..")
from models.utils_resnet import resnet18, resnet34, resnet50, resnet101, resnet152
from torchsummary import summary
```

```python
flag_gpu_available = torch.cuda.is_available()

if flag_gpu_available:
    device = torch.device("cuda")
else:
    device = torch.device("cpu")
```

```python
embedding_dimension = 256
```

### ResNet-18

```python
model = resnet18(pretrained=True)
```

```python
model.to(device)
```

```python
summary(model, input_size=(3, 224, 224))
```

```python
model = resnet18(pretrained=False)
```

```python
model.to(device)
```

```python
summary(model, input_size=(3, 224, 224))
```

```python
model.fc.in_features
```

```python
model.fc = nn.Sequential(
    nn.Linear(512, embedding_dimension, bias=False),
    nn.BatchNorm1d(embedding_dimension, eps=0.001, momentum=0.1, affine=True)
)
```

```python
model.fc
```

```python
model.to(device)
```

```python
summary(model, input_size=(3, 224, 224))
```
