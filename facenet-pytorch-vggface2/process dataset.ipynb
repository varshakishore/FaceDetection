```python
data_path = "../data/CASIA-WebFace/"

id_set = set([f for f in os.listdir(data_path) if not f.endswith("txt")])
id_list = list(id_set)

print(len(id_list))
```

```python
np.random.seed(0)
rand_perm = np.random.permutation( len(id_list) )

id_list_train = [id_list[i] for i in rand_perm[:int(len(id_list) * 0.9)]]
id_list_test = [id_list[i] for i in rand_perm[int(len(id_list) * 0.9):]]
```

```python
file = open('/home/rw565/FaceDetection/facenet-pytorch-vggface2/datasets/webface_train.csv', 'w') 
file.write("id,name,class\n")
for class_id, name in enumerate(id_list_train):
    img_id_set = set([f for f in os.listdir(data_path + name) if f.endswith("jpg")])
    img_id_list = list(img_id_set)
    for img_id in img_id_list:
        file.write(','.join([img_id.split(".")[0], name, str(class_id)]) + "\n")

file = open('/home/rw565/FaceDetection/facenet-pytorch-vggface2/datasets/webface_test.csv', 'w') 
file.write("id,name,class\n")
for class_id, name in enumerate(id_list_test):
    img_id_set = set([f for f in os.listdir(data_path + name) if f.endswith("jpg")])
    img_id_list = list(img_id_set)
    for img_id in img_id_list:
        file.write(','.join([img_id.split(".")[0], name, str(class_id)]) + "\n")

```
